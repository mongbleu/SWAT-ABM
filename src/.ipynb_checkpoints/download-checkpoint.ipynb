```python
from isimip_client.client import ISIMIPClient
import os
```

```python
client = ISIMIPClient()

# search the ISIMIP repository using a search string
response = client.datasets(query='gfdl-esm4 ssp370 pr')

# search the ISIMIP repository for a specific subtree
simulation_round = "ISIMIP3b"
product = "InputData"
scenario = "historical"
model_name = "ipsl-cm6a-lr"
climate_forcing = "gfdl-esm4"
climate_vars = ["hurs", "huss", "pr", "ps", "rids", "rsds", "sfcwind", "tasmax", "tasmin"]
response = client.datasets(tree=rf'ISIMIP3b/InputData/climate/atmosphere/global/daily/{scenario}/{model_name}/r1i1p1f1/w5e5/{climate_var}')

# search the ISIMIP repository using specifiers
response = client.datasets(simulation_round=simulation_round,
                           product=product,
                           climate_forcing=climate_forcing,
                           climate_scenario=scenario,
                           climate_variable=climate_var)
```

```python
datasets = response['results']
```

```python
parent_dir = r"E:\3.0.Code\CWatM-jupyter\data"
folder_name = climate_forcing + "-" + scenario
save_dir = os.path.join(parent_dir, folder_name)
if not os.path.exists(save_dir):
    print("No such directory exists.")
    os.mkdir(save_dir)
```

```python
# 용량 확인
total_size = 0
for ds in datasets:
    for f in ds["files"]:
        size_gb = f["size"] / 1e9
        total_size += size_gb
        print(f["name"], f"{size_gb:.2f} GB")

print("Total:", f"{total_size:.1f} GB")
```

```python
os.chdir(save_dir)
print(os.getcwd())
for ds in datasets:
    for f in ds['files']:
        client.download(f["file_url"], path=".")
```
